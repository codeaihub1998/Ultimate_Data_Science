```python
#Using Pickling approach-1
import pickle
import numpy as np
# Create three NumPy ndarray objects
array1 = np.array([[1, 2, 3], [4, 5, 6]])
array2 = np.array([[7, 8, 9], [10, 11, 12]])
array3 = np.array([[13, 14, 15], [16, 17, 18]])
```

```python
# Save the ndarray objects to a binary file using Pickle
with open('arrays_pickle.bin', 'wb') as f:
 pickle.dump(array1, f)
 pickle.dump(array2, f)
 pickle.dump(array3, f)
# Read and display ndarray objects from the file
with open('arrays_pickle.bin', 'rb') as f:
 loaded_array1 = pickle.load(f)
 loaded_array2 = pickle.load(f)
 loaded_array3 = pickle.load(f)
print("Loaded Array 1:")
print(loaded_array1)
print("\nLoaded Array 2:")
print(loaded_array2)
print("\nLoaded Array 3:")
print(loaded_array3)
```

```python
# Using NumPy's save and load functions approach-2
import numpy as np
# Create three NumPy ndarray objects
array1 = np.array([[1, 2, 3], [4, 5, 6]])
array2 = np.array([[7, 8, 9], [10, 11, 12]])
array3 = np.array([[13, 14, 15], [16, 17, 18]])
# Save the ndarray objects to a binary file using NumPy's save function
np.savez('arrays_npz.npz', array1=array1, array2=array2, array3=array3)
# Read and display ndarray objects from the file using NumPy's load function
loaded_data = np.load('arrays_npz.npz')

loaded_array1 = loaded_data['array1']
loaded_array2 = loaded_data['array2']
loaded_array3 = loaded_data['array3']
print("Loaded Array 1:")
print(loaded_array1)
print("\nLoaded Array 2:")
print(loaded_array2)
print("\nLoaded Array 3:")
print(loaded_array3)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.interpolate import interp1d

# Generate sample data representing a time series
time = np.array([0, 1, 2, 3, 4, 5, 6])
temperature = np.array([20, 22, 25, 24, 23, 22, 21])
# Perform linear interpolation
f_linear = interp1d(time, temperature)
# Generate denser time points for plotting the interpolated curve
time_dense = np.linspace(0, 6, 100)
temperature_dense_linear = f_linear(time_dense)
# Plot original data points and interpolated curve
plt.figure(figsize=(8, 6))
plt.scatter(time, temperature, label='Original Data', color='blue')
plt.plot(time_dense, temperature_dense_linear,
label='Linear Interpolation', color='orange')
plt.title('Temperature Interpolation')
plt.xlabel('Time (hours)')
plt.ylabel('Temperature (°C)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Sample data
x = np.arange(1, 6)
y_line = np.array([3, 5, 7, 4, 6])
y_bar = np.array([8, 6, 9, 7, 5])

# Create a figure and axis object
fig, ax = plt.subplots()

# Plot the line plot
ax.plot(x, y_line, marker='o', linestyle='-', color='b', label='Line Plot')

# Plot the bar plot
ax.bar(x, y_bar, color='r', alpha=0.5, label='Bar Plot')

# Set labels and title
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.set_title('Line and Bar Plot')

# Add legend
ax.legend()

# Show plot
plt.show()
```

```python
import pandas as pd
import polars as pl
# Creating a DataFrame/Series
# Pandas
df_pandas = pd.DataFrame({'A': [1, 2, 3, 4, 5], 'B': [6, 7, 8, 9, 10]})
series_pandas = pd.Series([1, 2, 3, 4, 5])
# Polars
df_polars = pl.DataFrame({'A': [1, 2, 3, 4, 5], 'B': [6, 7, 8, 9, 10]})
series_polars = pl.Series([1, 2, 3, 4, 5])

# Accessing data
# Pandas DataFrame
print("Pandas DataFrame:")
print(df_pandas)

# Polars DataFrame
print("\nPolars DataFrame:")
print(df_polars)

# Pandas Series
print("\nPandas Series:")
print(series_pandas)


# Polars Series
print("\nPolars Series:")
print(series_polars)
# Modifying data
# Pandas
df_pandas['C'] = df_pandas['A'] + df_pandas['B']
series_pandas = series_pandas + 1
# Polars
df_polars = df_polars.with_columns([series_polars.rename("C")])
series_polars = series_polars + 1

# Display modified data
print("\nModified Pandas DataFrame:")
print(df_pandas)
print("\nModified Polars DataFrame:")
print(df_polars)
print("\nModified Pandas Series:")
print(series_pandas)
print("\nModified Polars Series:")
print(series_polars)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
# Sample data
np.random.seed(0)
x = np.random.randn(100)
y = np.random.randn(100)
# Scatter plot using Seaborn
plt.figure(figsize=(6, 3))
sns.scatterplot(x=x, y=y, color='blue', label='Seaborn')
plt.title('Seaborn Scatter Plot') # Change title for Seaborn plot
plt.xlabel('X-axis')
plt.ylabel('Y-axis')

# Scatter plot using Matplotlib
plt.figure(figsize=(6, 3))
plt.scatter(x, y, color='red', label='Matplotlib')
plt.title('Matplotlib Scatter Plot') # Change title for Matplotlib plot
plt.xlabel('X-axis')
plt.ylabel('Y-axis')

# Show plots
plt.show()
```
